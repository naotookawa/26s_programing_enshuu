# week9 提出課題

## 課題の要約

郵便番号の文字列から対応する URL の JSON を取得し、住所を表示する。今回は授業資料の `showaddress` 問題を選び、HTTP と JSON を組み合わせて実装する。

## 実装方針

`urllib.request.urlopen` で JSON を取得し、`json.load` で Python の値に変換する。JSON の形式が練習用 URL と実データで少し違っても対応できるよう、よく使われる住所用フィールド名を順に探して表示する。実行確認はネットワーク依存を避けるため、ローカル JSON ファイルでもできるようにした（第9章 9.3「HTTP」, p.163-164; 第9章 9.7「この章の提出課題」, p.171）。

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path
import json

spec = spec_from_file_location('ch09_answer', Path('..') / 'outputs' / 'ch09' / 'answer.py')
ch09_answer = module_from_spec(spec)
assert spec is not None and spec.loader is not None
spec.loader.exec_module(ch09_answer)

extract_address = ch09_answer.extract_address
showaddress_from_file = ch09_answer.showaddress_from_file

def test_extract_address():
    assert extract_address({'prefecture': '東京都', 'city': '目黒区', 'town': '大橋'}) == '東京都目黒区大橋'
    assert extract_address({'allAddress': '東京都目黒区大橋'}) == '東京都目黒区大橋'
    assert extract_address({'results': [{'address': '東京都目黒区'}]}) == '東京都目黒区'

test_extract_address()
print('extract_address tests passed')
```

```python
sample_path = Path('..') / 'outputs' / 'ch09' / 'sample_postal.json'
sample_path.parent.mkdir(parents=True, exist_ok=True)
sample_path.write_text(json.dumps({'prefecture': '東京都', 'city': '目黒区', 'town': '大橋'}), encoding='utf-8')

address = showaddress_from_file(str(sample_path))
print('address =', address)
assert address == '東京都目黒区大橋'
```

## 実行結果

JSON から住所文字列を取り出せることを確認した。ローカルファイル経由の検証では `東京都目黒区大橋` が表示される。実運用では `showaddress(zipcode)` が URL から JSON を取得して表示する。
