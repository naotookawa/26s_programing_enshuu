# week7 提出課題

## 課題の要約

第7章のスターなし問題をすべて Python で解く。内部状態を持つクラス、メソッドによる再帰、メモ化、囚人のジレンマのプレイヤーと繰り返しゲームを実装する。

## 実装方針

銀行口座とカウンターは内部状態をクラスに閉じ込め、`ClockTimer` は秒の繰り上がりをそのまま実装する。`Fib` 系は再帰とカウンター、`MemoFib` は配列メモ化で高速化した。囚人のジレンマでは `Player` を共通インターフェースとして用意し、`TitForTatPlayer` は前回の相手行動を内部状態に保持する（第7章 7.1-7.3, p.115-132）。

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path

spec = spec_from_file_location('ch07_answer', Path('..') / 'outputs' / 'ch07' / 'answer.py')
ch07_answer = module_from_spec(spec)
assert spec is not None and spec.loader is not None
spec.loader.exec_module(ch07_answer)

BankAccount0 = ch07_answer.BankAccount0
BankAccount = ch07_answer.BankAccount
BankAccount2 = ch07_answer.BankAccount2
Counter = ch07_answer.Counter
ClockTimer = ch07_answer.ClockTimer
Fib = ch07_answer.Fib
CountedFib = ch07_answer.CountedFib
MemoFib = ch07_answer.MemoFib
Cooperate = ch07_answer.Cooperate
Defect = ch07_answer.Defect
CooperatePlayer = ch07_answer.CooperatePlayer
DefectPlayer = ch07_answer.DefectPlayer
RandomPlayer = ch07_answer.RandomPlayer
TitForTatPlayer = ch07_answer.TitForTatPlayer
play_one_game = ch07_answer.play_one_game
play_games = ch07_answer.play_games

def test_stateful_classes():
    a = BankAccount0(10000)
    b = BankAccount(10000)
    c = BankAccount2(10000, 'secret')
    assert a.withdraw(5000) == 5000
    assert b.withdraw(5000) == 5000
    assert b.deposit(4000) == 9000
    assert c.withdraw(4000, 'secret') == 6000
    assert c.deposit(5000, 'wrong') == 'Incorrect password'
    counter = Counter()
    counter.increment(); counter.increment()
    assert counter.count() == 2
    timer = ClockTimer(22, 54, 14)
    timer.tick()
    assert (timer.hour, timer.minute, timer.second) == (22, 54, 15)
    for _ in range(45):
        timer.tick()
    assert (timer.hour, timer.minute, timer.second) == (22, 55, 0)

test_stateful_classes()
print('stateful class tests passed')
```

```python
def test_fib_and_memo():
    f = Fib()
    assert f.fib(10) == 55
    g = CountedFib()
    assert g.fib(3) == 2
    assert g.count() > 0
    m = MemoFib()
    assert m.fib(10) == 55
    count1 = m.count()
    assert m.fib(10) == 55
    assert m.count() == count1

test_fib_and_memo()
print('fib / memo tests passed')
```

```python
import random

random.seed(0)
pc = CooperatePlayer()
pd = DefectPlayer()
pr = RandomPlayer()
tft = TitForTatPlayer()

assert tft.play() == Cooperate
tft.update(Cooperate, Defect)
assert tft.play() == Defect

play_one_game(pc, pd)
play_one_game(pc, pr)
play_one_game(pd, pr)

summary, history = play_games(pc, pr)
print('play_games summary =', summary)
print('history =', history)
assert len(summary) == 2
assert len(history) == 2
```
