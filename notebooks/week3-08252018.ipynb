## 課題の要約

第3章の提出課題として、通常問題の `score(a)` をループ不変条件つきで実装し、スター付き課題の `integral` をPythonで解く。`score` にはユニットテストを付け、`integral` は台形公式とシンプソン公式で近似する。

## 実装方針

`score(a)` は配列を1回走査して合計、最小値、最大値を更新し、最後に最小値と最大値を引く。ループの各時点で `total` がこれまでの合計、`current_min` と `current_max` がこれまでの最小値・最大値であることをコメントで明示した。`integral` は区間を等分して台形公式とシンプソン公式を実装し、二次関数と一次関数で精度確認を行った（第3章 3.4, p.52; 第3章 3.6, p.57-58; 第3章 3.7「この章の提出課題」, p.59）。

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path

spec = spec_from_file_location('ch03_answer', Path('..') / 'outputs' / 'ch03' / 'answer.py')
ch03_answer = module_from_spec(spec)
assert spec is not None and spec.loader is not None
spec.loader.exec_module(ch03_answer)

score = ch03_answer.score
integral_trapezoid = ch03_answer.integral_trapezoid
integral_simpson = ch03_answer.integral_simpson

def test_score():
    assert score([3, 2, 5]) == 3
    assert score([10, 1, 3, 7]) == 10
    assert score([8, 8, 8]) == 8
    assert score([1.5, 2.5, 10.0, 4.0]) == 6.5

test_score()
print('score tests passed')
```

```python
def f_quad(x):
    return x * x

def f_line(x):
    return 2.0 * x + 1.0

trap = integral_trapezoid(f_quad, 0.0, 1.0, 1000)
simp = integral_simpson(f_quad, 0.0, 1.0, 1000)

print('trapezoid x^2 on [0, 1] =', trap)
print('simpson x^2 on [0, 1] =', simp)
print('simpson linear exact =', integral_simpson(f_line, 0.0, 2.0, 20))

assert abs(trap - (1.0 / 3.0)) < 1e-3
assert abs(simp - (1.0 / 3.0)) < 1e-10
assert abs(integral_simpson(f_line, 0.0, 2.0, 20) - 6.0) < 1e-12
```

## 実行結果

`score` のテストは通過し、`integral` は二次関数で近似値が真値 `1/3` に十分近く、一次関数ではシンプソン公式が正確に再現できることを確認した。
