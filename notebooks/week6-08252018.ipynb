## 課題の要約

第6章の 6.2 までの非スター問題を Python で解く。木の基本操作、二分探索木の構築、Q&A の状態表現、四則演算の構文木の評価を実装する。

## 実装方針

内部表現は章の前半で説明されている配列と dict を使い、`make_node` / `value` / `left` / `right` / `is_empty` を通して操作を統一した。再帰で木をたどる関数は、空木を基底ケースにして素直に定義している。`make_binary_search_tree` は `add_node` を順に適用し、Q&A は質問ノードと答えノードを dict で表現した（第6章 6.1, p.103-111; 第6章 6.2, p.111-112; 第6章 6.4「この章の提出課題」, p.114）。

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path

spec = spec_from_file_location('ch06_answer', Path('..') / 'outputs' / 'ch06' / 'answer.py')
ch06_answer = module_from_spec(spec)
assert spec is not None and spec.loader is not None
spec.loader.exec_module(ch06_answer)

make_node = ch06_answer.make_node
make_leaf = ch06_answer.make_leaf
count_node = ch06_answer.count_node
max_value = ch06_answer.max_value
depth = ch06_answer.depth
preorder = ch06_answer.preorder
inorder = ch06_answer.inorder
make_binary_search_tree = ch06_answer.make_binary_search_tree
make_question_node = ch06_answer.make_question_node
question_text = ch06_answer.question_text
question_yes = ch06_answer.question_yes
question_no = ch06_answer.question_no
make_answer_node = ch06_answer.make_answer_node
answer_text = ch06_answer.answer_text
is_question_node = ch06_answer.is_question_node
calc = ch06_answer.calc

def test_tree_basics():
    t0 = make_leaf(2)
    t1 = make_node(1, make_leaf(3), make_leaf(5))
    t2 = make_node(3, t1, t0)
    assert count_node(t0) == 1
    assert count_node(t1) == 3
    assert count_node(t2) == 5
    assert max_value(t2) == 5
    assert depth(t0) == 1
    assert depth(t1) == 2
    assert preorder(t1) == [1, 3, 5]
    assert inorder(t1) == [3, 1, 5]

test_tree_basics()
print('tree basics tests passed')
```

```python
bst = make_binary_search_tree([3, 1, 6, 1, 7, 9, 2, 0, 5])
assert inorder(bst) == [0, 1, 1, 2, 3, 5, 6, 7, 9]

q = make_question_node(
    'Hungry',
    make_answer_node('cup noodle'),
    make_answer_node('take a nap'),
)
assert is_question_node(q) is True
assert question_text(q) == 'Hungry'
assert answer_text(question_yes(q)) == 'cup noodle'
assert answer_text(question_no(q)) == 'take a nap'

expr = make_node('+', make_node('*', make_leaf(3), make_leaf(5)), make_leaf(2))
assert calc(expr) == 17

print('bst / q&a / calc tests passed')
print('inorder(bst) =', inorder(bst))
print('calc(expr) =', calc(expr))
```
