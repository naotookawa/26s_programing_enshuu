## 課題の要約

第8章のスターなし問題を Python で作成する。`help` と `doctest` を使える形で docstring を付け、継承と Observer パターンを含むクラス群を実装する。

## 実装方針

第8章前半は `BankAccount` と `Player` の抽象インターフェースを docstring 付きで実装し、後半は `Subject` / `Observer` と `ClockTimer` / `ConsoleClock` / `CountdownClock` を組み合わせた。`MultiClock` を追加して、複数 observer の振る舞いも確認できるようにした（第8章 8.2, p.137-138; 第8章 8.3, p.139-143; 第8章 8.4, p.144-147; 第8章 8.6「この章の提出課題」, p.154）。

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path

spec = spec_from_file_location('ch08_answer', Path('..') / 'outputs' / 'ch08' / 'answer.py')
ch08_answer = module_from_spec(spec)
assert spec is not None and spec.loader is not None
spec.loader.exec_module(ch08_answer)

rectarea = ch08_answer.rectarea
BankAccount = ch08_answer.BankAccount
Player = ch08_answer.Player
Cooperate = ch08_answer.Cooperate
Defect = ch08_answer.Defect
CooperatePlayer = ch08_answer.CooperatePlayer
DefectPlayer = ch08_answer.DefectPlayer
RandomPlayer = ch08_answer.RandomPlayer
play_one_game = ch08_answer.play_one_game
play_games = ch08_answer.play_games
ClockTimer = ch08_answer.ClockTimer
ConsoleClock = ch08_answer.ConsoleClock
MultiClock = ch08_answer.MultiClock
CountdownClock = ch08_answer.CountdownClock

def test_rectarea_and_account():
    assert rectarea(7, 5) == 35
    a = BankAccount(10000)
    assert a.withdraw(5000) == 5000
    assert a.deposit(4000) == 9000
    assert a.withdraw(6000) == 3000

test_rectarea_and_account()
print('rectarea and account tests passed')
```

```python
import random

random.seed(0)
assert CooperatePlayer().play() == Cooperate
assert DefectPlayer().play() == Defect
assert isinstance(RandomPlayer().play(), int)
print(play_one_game(CooperatePlayer(), DefectPlayer()))
print(play_games(CooperatePlayer(), DefectPlayer())[0])
print('player tests passed')
```

```python
timer = ClockTimer(22, 54, 12)
cc = ConsoleClock(timer)
mc = MultiClock(timer)
timer.tick()
timer.tick()
assert mc.history == [(22, 54, 13), (22, 54, 14)]
timer.detach(cc)
timer.tick()
assert mc.history[-1] == (22, 54, 15)

timer2 = ClockTimer(22, 54, 12)
cd = CountdownClock(timer2, 3)
timer2.tick()
timer2.tick()
assert cd.alarmed is False
timer2.tick()
assert cd.alarmed is True
print('observer tests passed')
```
