## 課題の要約

第2章の `calcvar` は、変数 `x` を含む四則演算の式を文字列で受け取り、その式を値に評価する関数 `parse(s)` を作る課題である。

## 実装方針

`x` と整数を末端、`+ - * /` と括弧を内部節として扱う再帰下降で解く。`eval` は使わず、部分式を関数に変換して合成する。資料の関数と計算の考え方に合わせて、`make_term`・`calculator`・`parse_term`・`parse` を分けて実装する（第2章 2.2.4, p.24-25; 第2章 2.4「発展: 関数を返す関数」, p.34-38; 第2章 2.5「この章の提出課題」, p.39）。

```python
def make_term(token: str):
    if token == 'x':
        return lambda x: x
    value = int(token)
    return lambda x: value


def calculator(a: int, op: str, b: int) -> int:
    if op == '+':
        return a + b
    if op == '-':
        return a - b
    if op == '*':
        return a * b
    if op == '/':
        return a // b
    raise ValueError(f'unsupported operator: {op}')


def parse_one(s: str):
    lf = make_term(s[0])
    op = s[1]
    rf = make_term(s[2])
    return lambda x: calculator(lf(x), op, rf(x))


def parse_term(s: str):
    if s[0] == '(':
        lf, lsize = parse_term(s[1:])
        op_index = 1 + lsize
        op = s[op_index]
        rf, rsize = parse_term(s[op_index + 1:])
        close_index = op_index + 1 + rsize

        def func(x):
            return calculator(lf(x), op, rf(x))

        return func, close_index + 1

    i = 0
    while i < len(s) and s[i] not in '+-*/()':
        i += 1
    return make_term(s[:i]), i


def parse(s: str):
    f, size = parse_term(s)
    if size != len(s):
        raise ValueError('unexpected trailing characters')
    return f


tests = [
    ('((x+3)+3)', [0, 1]),
    ('((x+3)*(x-3))', [1]),
]
for expr, xs in tests:
    f = parse(expr)
    print(expr)
    for x in xs:
        print(x, '->', f(x))
```

## 実行結果

式文字列を関数に変換でき、`x` の値を変えても同じ式を再利用できる。`parse("((x+3)+3)")` と `parse("((x+3)*(x-3))")` の確認も通している。
