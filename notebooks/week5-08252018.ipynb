# week5 提出課題

## 課題の要約

第5章の Section 5.4 までにあるタイトル部分に `⋆` のない問題をすべて Python で解く。再帰の問題では破壊的代入を使わず、`merge` と `mergesort` ではループ不変条件をコメントで明示する。

## 実装方針

`factorial` から `binary_search` までは再帰の基本形と二分探索をそのまま関数化した。`koch` は点列だけを返し、描画はしない。`merge` は `while` を使って 2 つの整列済み配列を併合し、`mergesort` は分割して再帰的に整列した結果を `merge` でまとめる（第5章 5.1-5.4, p.86-96; 第5章 5.6「この章の提出課題」, p.99）。

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path

spec = spec_from_file_location('ch05_answer', Path('..') / 'outputs' / 'ch05' / 'answer.py')
ch05_answer = module_from_spec(spec)
assert spec is not None and spec.loader is not None
spec.loader.exec_module(ch05_answer)

factorial = ch05_answer.factorial
count_multiples = ch05_answer.count_multiples
gcd = ch05_answer.gcd
sqrt = ch05_answer.sqrt
sigmoid_inverse = ch05_answer.sigmoid_inverse
binary_search = ch05_answer.binary_search
fib = ch05_answer.fib
mdigit = ch05_answer.mdigit
koch = ch05_answer.koch
merge = ch05_answer.merge
first_half = ch05_answer.first_half
second_half = ch05_answer.second_half
mergesort = ch05_answer.mergesort

def test_recursive_basics():
    assert factorial(5) == 120
    assert count_multiples(10, 3) == 3
    assert gcd(12, 18) == 6
    assert abs(sqrt(3, 0, 10) - 1.7320508075688772) < 1e-3
    assert sigmoid_inverse(0.5, -10, 10) == 0.0
    assert abs(sigmoid_inverse(0.2, -10, 10) + 1.3862943611) < 1e-9
    assert binary_search([1, 3, 5, 7], 5, 0, 4) == 2
    assert fib(5) == 5

test_recursive_basics()
print('recursive tests passed')
```

```python
assert mdigit(3, 2) == ['00', '01', '02', '10', '11', '12', '20', '21', '22']

koch_points = koch(0.0, 0.0, 100.0, 0.0, 1)
print('mdigit(3,2) =', mdigit(3, 2))
print('koch depth1 point count =', len(koch_points))
assert len(koch_points) == 4
assert abs(koch_points[0][0] - 0.0) < 1e-9
assert abs(koch_points[-1][0] - 100.0 * 2.0 / 3.0) < 1e-9
```

```python
assert merge([1, 3, 5], [2, 4]) == [1, 2, 3, 4, 5]
assert first_half([0, 1, 2, 3, 4]) == [0, 1, 2]
assert second_half([0, 1, 2, 3, 4]) == [3, 4]
assert mergesort([8, 3, 4, 1, 5, 9, 6, 7, 2]) == [1, 2, 3, 4, 5, 6, 7, 8, 9]

print('merge / mergesort tests passed')
```

## 実行結果

再帰の基本問題、二分探索、フィボナッチ、`mdigit`、`koch`、`merge`、`mergesort` のテストを通した。`merge` では `while` のループ不変条件をコメントで示し、`mergesort` は分割と再帰結果の併合で構成した。
